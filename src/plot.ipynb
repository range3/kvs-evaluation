```python
import sys
sys.path.insert(0, "..")
from pathlib import Path
import re
import json
from pprint import pprint
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.collections as mc
import matplotlib.cm as cm
```

```python
df = (
    pd.read_csv(
        'result.csv',
        names=('workload', 'db', 'nthreads', 'KTPS'),
        dtype={'nthreads': int, 'KTPS': float},
    )
)
```

```python
workloads = df['workload'].unique()
```

```python
workloads = df['workload'].unique()
for workload in workloads:
    fig = plt.figure(figsize=(8,6), dpi=100)
    fig.patch.set_alpha(1)
    ax = fig.add_subplot()
    df[df['workload'] == workload].pivot(index='nthreads', columns=['workload', 'db'], values=['KTPS']).plot(ax=ax, kind='bar')
    ax.set_ylabel('KTPS')
    if workload != 'workloade.spec':
        ax.set_ylim(0, 2200)
```
